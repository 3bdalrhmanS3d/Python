## Data Analayis

### Questions 


1. **First Payment:**
   - Does a higher first installment affect loan approval?
   - What is the relationship between the first installment and loan rejection rates?

2. **Credit Amount:**
   - To what extent does the credit amount influence the loan approval decision?
   - Are smaller loans more likely to be approved than larger loans?

3. **Credit Term:**
   - Are longer-term loans more likely to be rejected compared to shorter-term loans?

4. **Income Norm (Normalized Income):**
   - Do customers with lower normalized income tend to have their loans rejected more?

5. **Income Sum:**
   - To what extent does the total income of the customer influence the approval decision?

6. **Confirmed Income:**
   - Is there a difference in loan approval decisions based on confirmed income versus declared income?

7. **Monthly Payment:**
   - Does a higher monthly payment lead to an increased rejection rate?

8. **Gender:**
   - Is there a difference in loan approvals between male and female customers?

9. **Family Children (Number of Children):**
   - Do customers with more children tend to have their loans rejected more often?

10. **Marriage Status:**
    - Are married customers more likely to have their loan approved or rejected?

11. **Job Experience (Current Job Experience):**
    - Do customers with longer job experience in their current job tend to have their loans approved more?

12. **General Job Experience:**
    - To what extent does overall job experience impact loan approval?

13. **Family Years (Years of Marriage):**
    - Do customers with longer years of marriage tend to have their loans approved?

14. **Education Level:**
    - Does a higher level of education positively influence loan approval?

15. **Age:**
    - Are older customers more likely to have their loans rejected?

16. **Official Job Status:**
    - Are customers with official jobs more likely to be approved for loans?

17. **Job Position:**
    - Are customers in managerial or leadership positions more likely to be approved than those in lower-level positions?

18. **Loan Decision:**
    - What features are the most influential in predicting loan approval or rejection?
